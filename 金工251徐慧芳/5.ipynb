```r
earl.posts <- r.posts[order(create_time), .SD[1], by = guba_name]
earl.posts
```

```r
late.replys <- r.replys[order(reply_time), .SD[.N], by = post_id]
late.replys
```

```r
r.replys[, reply_content := str_remove_all(reply_content, "<[^>]+>")]
int.data <- merge(r.replys, r.posts, by = "post_id", all = FALSE)
late.replys
```

```r
int.data[, create_date := as.Date(create_time, format = "%Y-%m-%d")]
int.data[, create_time := as.POSIXct(create_time, format = "%Y-%m-%d %H:%M:%OS")]
int.data
```

```r
date_range <- range(int.data$create_date, na.rm = TRUE)
all_dates <- seq(date_range[1], date_range[2], by = "day")
all_guba <- unique(int.data$guba_name)
tim <- expand.grid(guba_name = all_guba, create_date = all_dates)
setDT(tim)
tim
```

```r
int.data.ts <- merge(int.data, tim, by = c("guba_name", "create_date"), all.y = TRUE)
int.data.ts[, c("readnum", "replynum") := lapply(.SD, function(x) replace(x, is.na(x), 0)), .SDcols = c("readnum", "replynum")]
int.data.ts
```

```r
mx.post <- r.posts[order(-replynum)][1:10, .(post_id, guba_name, readnum, replynum)]
mx.ba <- r.posts[, .(total_read = sum(readnum), total_reply = sum(replynum)), by = guba_name][order(-total_reply)][1:10]
mx.post
mx.ba
```
